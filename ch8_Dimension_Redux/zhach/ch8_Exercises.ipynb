### Exercise 1
What are the main motivations for reducing a dataset's dimensionality? What are the main drawbacks?

Remove dimensions helps simplify the problem of learning by removing unnecessary/similarly correlated features. It speeds of training. Removes noise. And you can visualize it easier. 

However, it does remove some information, sometimes reducing the performance of the model. It can also be computationally intesive. Not to mention that transformed features are sometimes hard to interpret.

### Exercise 2
What is the curse of dimensionality?

The curse of dimensionality referes to the fact that many problems that do not exist in low-dimensional splace arise in high-dimensional spaces. One common problem is the fact that randomly sampled high-dimensional vectors are generally very sparse (far apart), which increases overfitting.

### Exercise 3
Once a dataset's dimensionality has been reduced, is it possible to reverse the operation? If so, how? If not, why?

It depends on the algorithm. PCA has the abilty to do a reverse transformation that can get a similar result as the original dataset. But algorithms like t-SNE do not.

### Exercise 4

Usually yes. But if you have a dataset that relies on all the features, reducing the dimensions can lose a lot of information.

### Exercise 8
Does it make any sense to chain two different dimensionality reduction algorithms?

Yes! Dim-redux algorithms have varying ranges of process times and dimensionality reduction capabilities. Lik eyou can run a PCA to quickly get a lower-dimension, then run a more complicated dim-redux algorithm to get better performance. 

```python
# Exercise 9
## Load the MNIST datatset (train/test)
## Train a Random Forest Classifier, time the train, and evalute with the test.
## Use PCA to reduce the dimensions, with 95% explained variance.
## Train a new Random Forest Classifier, time it, and evalute. 
## Was it faster than the previous model train? Did it perform better on the test set?
import numpy as np
from sklearn.datasets import fetch_mldata
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score, cross_val_predict

mnist = fetch_mldata("MNIST original")
X,y = mnist["data"], mnist["target"]
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
shuffle_index = np.random.permutation(60000)
X_train, y_train = X_train[shuffle_index], y_train[shuffle_index]
print "Data in memory"

for_clf1 = RandomForestClassifier(random_state=42)
for_clf1.fit(X_train, y_train)

y_test_knn_pred = cross_val_predict(for_clf1, X_test, y_test, cv=3)
normal_acc = cross_val_score(for_clf1, X_test, y_test, cv=3, scoring="accuracy").mean()
normal_f1 = f1_score(y_test, y_test_knn_pred, average="macro")
print "Normal Accuracy: ", normal_acc
print "Normal F1: ", normal_f1

pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X_train)
print "X_train has been reduced"
print X_train.shape
print X_reduced.shape


for_clf2 = RandomForestClassifier(random_state=42)
for_clf2.fit(X_reduced, y_train)

y_test_knn_pred = cross_val_predict(for_clf2, X_test, y_test, cv=3)
pca_acc = cross_val_score(for_clf2, X_test, y_test, cv=3, scoring="accuracy").mean()
pca_f1 = f1_score(y_test, y_test_knn_pred, average="macro")
print "PCA Accuracy: ", pca_acc
print "PCA F1: ", pca_f1

print "Accuracy change: ", pca_acc-normal_acc
print "F1 change: ", pca_f1-normal_f1
```

```python
# Exercise 10
## Use t-SNE to reduce the MNIST down to 2-dim and plot using Matplotlib (Scatterplot with 10 colors for numbers)
## Try other dim-redux algorithms to see different clusterings.
from sklearn.manifold import TSNE, LocallyLinearEmbedding, MDS
import matplotlib.pyplot as plt

mnist = fetch_mldata("MNIST original")
shuffle_index = np.random.permutation(60000)
m = 10000
idx = shuffle_index[:m]

X = mnist['data'][idx]
y = mnist['target'][idx]

tsne = TSNE(n_components=2, random_state=42)
X_reduced = tsne.fit_transform(X)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()

pca = PCA(n_components=2, random_state=42)
X_reduced = pca.fit_transform(X)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()

lle = LocallyLinearEmbedding(n_components=2, random_state=42)
X_reduced = lle.fit_transform(X)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()

mds = MDS(n_components=2, random_state=42)
X_reduced = mds.fit_transform(X)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()
```
